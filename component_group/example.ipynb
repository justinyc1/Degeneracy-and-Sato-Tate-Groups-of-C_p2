# Computing $\text{ST}(\text{Jac}(\text{C}_{p^2}))/ \text{ST}^{0}(\text{Jac}(\text{C}_{p^2}))$

## Intro and Structure

In this notebook, we will demonstrate computing a generator for $\text{ST}(\text{Jac}(\text{C}_{p^2}))/ \text{ST}^{0}(\text{Jac}(\text{C}_{p^2}))$, the component group of $\text{ST}(\text{Jac}(\text{C}_{p^2}))$. 

We make use of the `component_group_gen()` function, found in the *component_group_generator.sage* file, which is based on the information in Section 2.3.1 and Definition 4.3. We will focus on computing component group generators for $p=5, \, 7$.

```python
def component_group_gen(p):
    r = IntegerModRing(p^2).unit_gens()
    g = (p^2 - 1)/2
    
    exponents = [a for a in range(1,g+1)]
    newExponents = [(int(r[0])*a % (p^2)) for a in exponents]
    
    groupActEndo = []
    for a in newExponents:
        if a in exponents:
            groupActEndo.append(a)
        else:
            n = a - (p^2)
            groupActEndo.append(n)
    
    for i in exponents:
        for j in groupActEndo:
            if i == j:
                print(f"{groupActEndo.index(j), exponents.index(i)} = I")
            elif i == abs(j):
                print(f"{groupActEndo.index(j), exponents.index(i)} = J")
```

This function will return all non-zero entries of the component group generator (for a prime $p$). The remaining entries are $2 \times 2$ zero matrices. As mentioned in the *component_group_generator.sage* file, ***all indexing starts at zero here, not one.***

**A miscellaneous note:** The interesting "staircase" appearance of the non-zero entries of $\gamma$ may be due to the ordering of differential 1-forms for $C_{p^2}$ (as described in Section 2.3.1).
